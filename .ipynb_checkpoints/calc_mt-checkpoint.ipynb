```python
data = pd.read_csv('../GenomeCRISPR_full05112017.csv')
```

```python
def calc_mt(row):
    seq = Seq(row['sequence'])
#     seq = Seq(row['sequence'].values[0])           # need this if sequence is an object?
    return mt.Tm_NN(seq)
```

```python
data['melting_temp'] = data.apply(lambda row: calc_mt(row), axis=1)
```
